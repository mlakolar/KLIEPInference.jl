# Calibration

```julia
using JLD, PyPlot
```

```julia
Î± = 0.05:0.05:0.95;
```

## `m = 25, sgn = -1`

```julia
file = jldopen("res/coverage_25_-1.jld", "r")
coverage = read(file, "coverage")
coverage = broadcast(/, coverage, 1000)
close(file)

figure(figsize = (10, 5))

subplot(2, 4, 1)
plot(1.0 .- Î±, 1.0 .- Î±, color="grey", linestyle=":", linewidth=.25)
plot(1.0 .- Î±, coverage[1, :], marker="o")
title("SparKLIE+1 full", size="xx-small")

subplot(2, 4, 2)
plot(1.0 .- Î±, 1.0 .- Î±, color="grey", linestyle=":", linewidth=.25)
plot(1.0 .- Î±, coverage[2, :], marker="o")
title("SparKLIE+1 lite w/o bias correction", size="xx-small")

subplot(2, 4, 3)
plot(1.0 .- Î±, 1.0 .- Î±, color="grey", linestyle=":", linewidth=.25)
plot(1.0 .- Î±, coverage[3, :], marker="o")
title("SparKLIE+1 lite w/  bias correction", size="xx-small")

subplot(2, 4, 4)
plot(1.0 .- Î±, 1.0 .- Î±, color="grey", linestyle=":", linewidth=.25)
plot(1.0 .- Î±, coverage[4, :], marker="o")
title("SparKLIE+1 zero", size="xx-small")

subplot(2, 4, 5)
plot(1.0 .- Î±, 1.0 .- Î±, color="grey", linestyle=":", linewidth=.25)
plot(1.0 .- Î±, coverage[5, :], marker="o")
title("SparKLIE+2 full", size="xx-small")

subplot(2, 4, 6)
plot(1.0 .- Î±, 1.0 .- Î±, color="grey", linestyle=":", linewidth=.25)
plot(1.0 .- Î±, coverage[6, :], marker="o")
title("SparKLIE+2 zero", size="xx-small")

subplot(2, 4, 7)
plot(1.0 .- Î±, 1.0 .- Î±, color="grey", linestyle=":", linewidth=.25)
plot(coverage[1, :], coverage[5, :], marker="o")
title("SparKLIE+1 vs SparKLIE+2 full", size="xx-small")

subplot(2, 4, 8)
plot(1.0 .- Î±, 1.0 .- Î±, color="grey", linestyle=":", linewidth=.25)
plot(coverage[4, :], coverage[6, :], marker="o")
title("SparKLIE+1 vs SparKLIE+2 zero", size="xx-small")

tight_layout()
```
