# Calibration

```julia
using KLIEPInference
using ProximalBase, CoordinateDescent
using LinearAlgebra, SparseArrays, Statistics, Random
using Distributions, StatsBase, PyPlot
```

```julia
function boot_SparKLIE(Ψx, Ψy, θ, Hinv; bootSamples::Int64=300)
    p, nx = size(Ψx)
    ny = size(Ψy, 2)
    supp = findall(!iszero, θ)

    θ1 = Vector{Float64}(undef, p)
    θ2 = Vector{Float64}(undef, p)

    bθ1 = Matrix{Float64}(undef, p, bootSamples)
    bθ2 = Matrix{Float64}(undef, p, bootSamples)

    x_ind = Matrix{Int16}(undef, nx, bootSamples)
    y_ind = Matrix{Int16}(undef, ny, bootSamples)

    for b = 1:bootSamples
        sample!(1:nx, view(x_ind, :, b))
        sample!(1:ny, view(y_ind, :, b))
    end

    for k = 1:p
        suppk = sort(union(supp, k))

        θk = SparseIterate(p)
        θk[suppk] = KLIEP(Ψx[suppk, :], Ψy[suppk, :], CD_KLIEP())

        θ1[k] = KLIEP_debias1(Ψx, Ψy, θk, Hinv[k], k)
        θ2[k] = KLIEP_debias2(Ψx, Ψy, θ, Hinv[k], k)

        bΨx = similar(Ψx)
        bΨy = similar(Ψy)

        for b = 1:bootSamples
            KLIEPInference._fill_boot_Psi!(bΨx, Ψx, view(x_ind, :, b))
            KLIEPInference._fill_boot_Psi!(bΨy, Ψy, view(y_ind, :, b))

            θk = SparseIterate(p)
            θk[suppk] = KLIEP(bΨx[suppk, :], bΨy[suppk, :], CD_KLIEP())

            bθ1[k, b] = KLIEP_debias1(bΨx, bΨy, θk, Hinv[k], k)
            bθ2[k, b] = KLIEP_debias2(bΨx, bΨy, θ, Hinv[k], k)
        end
    end

    return BootstrapEstimates(θ1, bθ1), BootstrapEstimates(θ2, bθ2)
end
```

```julia
m = 25
sgn = 0

# generate a graph
θx = chain(m, 5, 0.2, 0.4, sgn)
θy = copy(θx)

p = length(θx)
nx = 500
ny = 500
```

```julia
α = 0.05:0.05:0.95
nα = length(α)
res = zeros(Float64, 3, nα);
```

```julia
nrep = 10;
```

```julia
for rep = 1:nrep
    println("$(rep) / $(nrep)")
    
    # generate samples
    spl = IsingSampler(θx; thin=2000)
    X = rand(spl, nx)
    spl = IsingSampler(θy; thin=2000)
    Y = rand(spl, ny)
    
    Ψx = Ψising(X)
    Ψy = Ψising(Y)
    
    λ1 = 1.01 * quantile(Normal(), 1. - 0.05 / p)
    θ = spKLIEP(Ψx, Ψy, λ1, CD_KLIEP(); loadings=true)
    supp = findall(!iszero, θ)
    θ[supp] = KLIEP(Ψx[supp, :], Ψy[supp, :], CD_KLIEP())
    
    λ2 = sqrt(2. * log(p) / ny)
    H = KLIEP_Hessian(spzeros(Float64, p), Ψy)
    Hinv = Vector{SparseIterate{Float64}}(undef, p)
    for k = 1:p
        ω = Hinv_row(H, k, λ2)
        
        Sk = sort(union(k, findall(!iszero, ω)))
        Hk = view(H, Sk, Sk)
        δk = (Sk .== k)
        ω[Sk] = Hk\δk
        
        Hinv[k] = ω
    end
    
    boot1, boot2 = boot_SparKLIE(Ψx, Ψy, θ, Hinv)
    
    for i = 1:nα
        CI = simulCI(boot1, α[i])
        covered1 = all(0 .<= CI[:, 2]) * all(0 .>= CI[:, 1]) ? 1 : 0
        
        CI = simulCI(boot2, α[i])
        covered2 = all(0 .<= CI[:, 2]) * all(0 .>= CI[:, 1]) ? 1 : 0
        
        res[1, i] += covered1
        res[2, i] += covered2
        res[3, i] += (covered1 == covered2)
    end
end
```

```julia
coverage = broadcast(/, res, nrep);
```

```julia
figure(figsize = (15, 5))

subplot(1, 3, 1)
plot(1.0 .- α, 1.0 .- α, color="grey", linestyle=":", linewidth=.25)
plot(1.0 .- α, coverage[1, :], marker="o")
title("SparKLIE+1", size="xx-small")

subplot(1, 3, 2)
plot(1.0 .- α, 1.0 .- α, color="grey", linestyle=":", linewidth=.25)
plot(1.0 .- α, coverage[2, :], marker="o")
title("SparKLIE+2", size="xx-small")

subplot(1, 3, 3)
plot(coverage[3, :], marker="o")
title("SparKLIE+1 vs SparKLIE+2", size="xx-small")

tight_layout()
```
